#### 3.1 UNIVARIATE ANALYSIS

#### 3.2 BIVARIATE ANALYSIS

The purpose of bivariate analysis is to study relationships between predictor variables and the target variable (Diabetes_012).

This helps identify strong diabetes indicators and relevant features for model training.

##### Continuous Features

```python


fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes = axes.flatten()

CLASS_PALETTE = ['#378ADD', '#EF9F27', '#E24B4A']
CLASS_LABELS  = {0.0: 'No diabetes', 1.0: 'Prediabetes', 2.0: 'Diabetes'}

for i, col in enumerate(['BMI', 'MentHlth', 'PhysHlth']):
    for cls, color in zip([0.0, 1.0, 2.0], CLASS_PALETTE):
        subset = df[df['Diabetes_012'] == cls][col]
        sns.kdeplot(subset, ax=axes[i], color=color,
                    linewidth=2, fill=True, alpha=0.25,
                    label=CLASS_LABELS[cls])

    axes[i].set_xlim(df[col].min(), df[col].quantile(0.99))
    axes[i].set_title(f'{col} by class', fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')
    axes[i].legend(fontsize=9)

plt.suptitle('Continuous Feature distribution by diabetes class',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

##### Ordinal Features

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
axes = axes.flatten()

CLASS_PALETTE = ['#378ADD', '#EF9F27', '#E24B4A']
CLASS_LABELS  = {0.0: 'No diabetes', 1.0: 'Prediabetes', 2.0: 'Diabetes'}

ordinal_cols = ['GenHlth', 'Age', 'Education', 'Income']

for i, col in enumerate(ordinal_cols):
    for cls, color in zip([0.0, 1.0, 2.0], CLASS_PALETTE):
        subset = df[df['Diabetes_012'] == cls][col]

        sns.kdeplot(
            subset,
            ax=axes[i],
            color=color,
            linewidth=2,
            fill=True,
            alpha=0.25,
            label=CLASS_LABELS[cls]
        )

    axes[i].set_title(f'{col} vs Diabetes', fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')
    axes[i].legend(fontsize=9)

plt.suptitle('Ordinal features distribution by diabetes class',
             fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

##### Binary Features

```python

# Simple comparison: Which group has more diabetes?

# Binary columns to compare
features = [
    'HighBP', 'HighChol', 'Smoker', 'Stroke',
    'HeartDiseaseorAttack', 'PhysActivity',
    'Fruits', 'Veggies', 'DiffWalk'
]

results = []

for col in features:
    # Diabetes rate among people with value = 1
    yes_rate = (df[df[col] == 1]['Diabetes_012'] == 2).mean() * 100
    
    # Diabetes rate among people with value = 0
    no_rate = (df[df[col] == 0]['Diabetes_012'] == 2).mean() * 100
    
    results.append([col, yes_rate, no_rate])

# Create dataframe
compare_df = pd.DataFrame(results, columns=['Feature', 'Yes(1)%', 'No(0)%'])
compare_df.set_index('Feature', inplace=True)

# Plot (same style)
compare_df.plot(kind='bar', figsize=(12,6), color=['red','gray'])

plt.title("Diabetes % in Yes vs No Groups")
plt.ylabel("Diabetes Percentage")
plt.xticks(rotation=45)
plt.legend(['Yes', 'No'])
plt.show()
```

### 3.3 CORRELATION ANALYSIS

Correlation analysis helps measure linear relationships between numeric variables. It helps detect multicollinearity and redundant features.

```python
# Correlation heatmap (with correlation values)
corr = df.corr(numeric_only=True)

plt.figure(figsize=(16, 12))
sns.heatmap(
    corr,
    cmap='coolwarm',
    center=0,
    annot=True,
    fmt='.2f',
    annot_kws={'size': 10},
    square=True,
    linewidths=0.3,
    cbar_kws={'shrink': 0.8}
 )
plt.title('Feature Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.show()
```
